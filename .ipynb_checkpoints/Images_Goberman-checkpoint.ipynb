```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
old_path = '../../Downloads/Д. Гоберман'
new_path = '../../Downloads/Goberman_new'

if not os.path.exists(new_path):
    os.makedirs(new_path)

folders = [f for f in os.listdir(path=old_path) if not f.startswith('.')]

for folder in folders:
    f_list = [folder for folder in os.listdir(path=old_path+'/'+folder) if not folder.startswith('.')]
    
    for sub_folder in f_list:
        im_list = [file for file in os.listdir(path=old_path+'/'+folder+'/'+sub_folder) if (not file.startswith('.')) and (not file == 'Thumbs.db') ]
        
        for im in im_list:
            new_im = sub_folder + '_' + im.lstrip('0')
            src_dir = old_path+'/'+folder+'/'+sub_folder+'/'+im
            dst_dir = new_path+'/'+new_im
            shutil.copy(src_dir,dst_dir)
            
            
```
